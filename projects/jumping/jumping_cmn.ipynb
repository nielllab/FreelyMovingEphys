```python
njumps = len(df)
szs = np.unique(df['distance'])
```

```python
dist = np.zeros(njumps,)
takeX0 = np.zeros(njumps,)
takeY0 = np.zeros(njumps,)
landX0 = np.zeros(njumps,)
landY0 = np.zeros(njumps,)

plt.figure(figsize = (8,20))
for jmp in range(njumps):
    jumpT = df['Side_Jump'].iloc[jmp]-df['Side_Start'].iloc[jmp]
    endT = df['Side_End'].iloc[jmp]-df['Side_Start'].iloc[jmp]

    noseX = df['Side Nose x'].iloc[jmp]
    noseY = df['Side Nose y'].iloc[jmp]
    noseP = df['Side Nose likelihood'].iloc[jmp]
    noseX[noseP<like_thresh] = np.nan;
    noseY[noseP<like_thresh] = np.nan;

    takeX = df['Side TakeFR x'].iloc[jmp]
    takeY = df['Side TakeFR y'].iloc[jmp]
    takeP = df['Side TakeFR likelihood'].iloc[jmp]
    takeX[takeP<like_thresh]= np.nan
    takeY[takeP<like_thresh]= np.nan
    takeX0[jmp] = np.nanmedian(takeX)
    takeY0[jmp] = np.nanmedian(takeY)

    landX = df['Side LandFR x'].iloc[jmp]
    landY = df['Side LandFR y'].iloc[jmp]
    landP =  df['Side LandFR likelihood'].iloc[jmp]
    landX[landP<like_thresh]= np.nan
    landY[landP<like_thresh]= np.nan
    landX0[jmp] = np.nanmedian(landX)
    landY0[jmp] = np.nanmedian(landY)

    #plt.plot(takeX,takeY,'g',linewidth = 4,label = 'takeoff')
   # plt.plot(landX,landY,'r',linewidth = 4, label = 'land')


    dist[jmp] = takeX0[jmp]-landX0[jmp];
    
    if (df['success'].iloc[jmp] == 2):
        for i in range(len(szs)):
            if df['distance'].iloc[jmp]==szs[i]:
                plt.subplot(len(szs),1,i+1)
    
        #x0 = noseX[jumpT-1]; y0 = noseY[jumpT-1]
        x0 - 500; y0=300
        plt.plot(noseX[:jumpT]-x0,noseY[:jumpT]-y0,'b',linewidth = 0.25, label = 'nose pre')
        plt.plot(noseX[jumpT:endT]-x0,noseY[jumpT:endT]-y0, 'k',linewidth = 0.25, label = 'nose post')
        plt.plot(noseX[endT-1]-x0,noseY[endT-1]-y0, 'g.',linewidth = 2)
        plt.plot(takeX0[jmp]-x0,takeY0[jmp]-y0,'r.',linewidth = 2,label = 'takeoff')
        plt.plot(landX0[jmp]-x0,landY0[jmp]-y0,'r.',linewidth = 2, label = 'land')

    for i in range(len(szs)):
        plt.subplot(len(szs),1,i+1)
        plt.xlabel('x - pix'); plt.ylabel('y -pix'); plt.xlim(-400,200); plt.ylim(150,-150)
        
#plt.legend(); 
#plt.title('dist=' + str(df['distance'].iloc[jmp]) + ' success=' + str(df['success'].iloc[jmp]))
plt.savefig('all_aborts.pdf')
```

```python
dist_types = [8, 12, 16, 20, 24]
success_types = [0, 1, 2]
eye_types = ['binocular', 'monocular']
performance = np.zeros((len(eye_types),len(dist_types),len(success_types)))
for d, dist in enumerate(dist_types):
    for s, success in enumerate(success_types):
        for e, eye in enumerate(eye_types):
            df_use = df[(df['laser_trial'] == 'laser off') &(df['distance']==dist) & (df['success']==success) & (df['ocular']==eye)]
            performance[e,d,s] = len(df_use)
for e in range(2):
    plt.subplot(1,2,e+1)
    plt.plot(dist_types,performance[e,:,:]/np.sum(performance[e,:,:],axis=1)[:,np.newaxis])
    plt.title(eye_types[e])
    plt.legend(['failure','success','abort']); plt.xlabel('dist'); plt.ylabel('fraction')
plt.tight_layout()
    
    
```

```python
alljumpsX = np.zeros((len(dist_types),len(success_types),len(eye_types),20))
alljumpsY = np.zeros((len(dist_types),len(success_types),len(eye_types),20))
                         
for d, dist in enumerate(dist_types):
    for s, success in enumerate(success_types):
        for e, eye in enumerate(eye_types):
            df_use = df[(df['laser_trial'] == 'laser off') &(df['distance']==dist) & (df['success']==success) & (df['ocular']==eye)]
            pdf = mkpdf.PdfPages('alltrials_'+eye+'_'+str(dist)+'_'+str(success) + '.pdf')
            jumpsX = np.zeros((20,len(df_use)))
            jumpsY = np.zeros((20,len(df_use)))
            for jmp in range(len(df_use)):
                plt.figure(figsize = (12,6))
                
                jumpT = df_use['Side_Jump'].iloc[jmp]-df_use['Side_Start'].iloc[jmp]
                endT = df_use['Side_End'].iloc[jmp]-df_use['Side_Start'].iloc[jmp]

                noseX = df_use['Side Nose x'].iloc[jmp]
                noseY = df_use['Side Nose y'].iloc[jmp]
                noseP = df_use['Side Nose likelihood'].iloc[jmp]
                noseX[noseP<like_thresh] = np.nan;
                noseY[noseP<like_thresh] = np.nan;
                jumpsX[:,jmp] = noseX[jumpT-10 : jumpT+10]
                jumpsY[:,jmp] = noseY[jumpT-10 : jumpT+10]
                
                landX = df_use['Side LandFR x'].iloc[jmp]
                landY = df_use['Side LandFR y'].iloc[jmp]
                landP =  df_use['Side LandFR likelihood'].iloc[jmp]
                landX[landP<like_thresh]= np.nan
                landY[landP<like_thresh]= np.nan
                landX0 = np.nanmedian(landX)
                landY0 = np.nanmedian(landY)
                    
                x0 - 500; y0=300
                        
                plt.plot(noseX[:jumpT]-x0,noseY[:jumpT]-y0,'b',linewidth = 0.25, label = 'nose pre')
                plt.plot(noseX[jumpT:endT]-x0,noseY[jumpT:endT]-y0, 'k',linewidth = 0.25, label = 'nose post')
                plt.plot(noseX[endT]-x0,noseY[endT]-y0, 'g.',linewidth = 2)
                plt.plot(landX0-x0,landY0-y0,'r.',linewidth = 2, label = 'land')
                    
                plt.xlim(-300,100); plt.ylim(100,-100)
                plt.title(df['subject'].iloc[jmp])
                pdf.savefig()
                plt.close()
            pdf.close()
            alljumpsX[d,s,e,:] = np.nanmean(jumpsX,axis = 1)
            alljumpsY[d,s,e,:] = np.nanmean(jumpsY,axis = 1)
            
                
            
```
